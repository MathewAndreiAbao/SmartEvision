## Section 4: Naive Bayes Model Training
Implementation of a Multinomial Naive Bayes classifier with Laplace smoothing.


```python
def tokenize(text):
    # Matches the app's internal tokenization logic
    text = text.lower()
    text = re.sub(r'[^a-z0-9\s]', '', text)
    tokens = text.split()
    return [t for t in tokens if len(t) > 2]

class NaiveBayesTrainer:
    def __init__(self):
        self.vocabulary = set()
        self.classes = {}
        self.V = 0

    def train(self, texts, labels):
        total_docs = len(texts)
        unique_labels = list(set(labels))
        
        # Build vocabulary
        for text in texts:
            for token in tokenize(text):
                self.vocabulary.add(token)
        
        self.V = len(self.vocabulary)
        vocab_list = list(self.vocabulary)
        
        model = {
            "classes": {},
            "vocabularySize": self.V
        }
        
        for label in unique_labels:
            label_texts = [texts[i] for i in range(total_docs) if labels[i] == label]
            label_doc_count = len(label_texts)
            prior = np.log(label_doc_count / total_docs)
            
            # Word counts for this class
            word_counts = {}
            total_words_in_class = 0
            for text in label_texts:
                tokens = tokenize(text)
                for token in tokens:
                    word_counts[token] = word_counts.get(token, 0) + 1
                    total_words_in_class += 1
            
            # Weighted probabilities with Laplace smoothing
            word_probs = {}
            for word in self.vocabulary:
                count = word_counts.get(word, 0)
                prob = np.log((count + 1) / (total_words_in_class + self.V))
                word_probs[word] = prob
                
            default_prob = np.log(1 / (total_words_in_class + self.V))
            
            model["classes"][label] = {
                "priorProbability": float(prior),
                "wordProbabilities": {k: float(v) for k, v in word_probs.items()},
                "defaultWordProb": float(default_prob)
            }
            
        return model

print("Trainer class defined.")
```

## Section 5: Model Evaluation & Metrics
Training and testing the models using an 80/20 split.


```python
# Train Subject Model
X_sub = df_subject['text'].values
y_sub = df_subject['label'].values
X_train_sub, X_test_sub, y_train_sub, y_test_sub = train_test_split(X_sub, y_sub, test_size=0.2, random_state=42)

trainer_sub = NaiveBayesTrainer()
sub_model = trainer_sub.train(X_train_sub, y_train_sub)

# Local Predictor for evaluation
def predict(model, text):
    tokens = tokenize(text)
    best_class = None
    best_score = -float('inf')
    
    for label, data in model['classes'].items():
        score = data['priorProbability']
        for token in tokens:
            if token in data['wordProbabilities']:
                score += data['wordProbabilities'][token]
            else:
                score += data['defaultWordProb']
        
        if score > best_score:
            best_score = score
            best_class = label
    return best_class

y_pred_sub = [predict(sub_model, text) for text in X_test_sub]

print("Subject Classifier Report:")
print(classification_report(y_test_sub, y_pred_sub))

plt.figure(figsize=(8, 6))
cm = confusion_matrix(y_test_sub, y_pred_sub)
sns.heatmap(cm, annot=True, fmt='d', xticklabels=np.unique(y_sub), yticklabels=np.unique(y_sub), cmap='Blues')
plt.title("Subject Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

## Section 6: Cross-Validation Reliability Assessment
Using 5-Fold Stratified Cross-Validation to ensure data consistency.


```python
skf = StratifiedKFold(n_splits=5)
scores = []

for train_index, test_index in skf.split(X_sub, y_sub):
    X_f_train, X_f_test = X_sub[train_index], X_sub[test_index]
    y_f_train, y_f_test = y_sub[train_index], y_sub[test_index]
    
    m = NaiveBayesTrainer().train(X_f_train, y_f_train)
    y_f_pred = [predict(m, text) for text in X_f_test]
    scores.append(accuracy_score(y_f_test, y_f_pred))

plt.figure(figsize=(8, 4))
plt.bar(range(1, 6), scores, color='skyblue')
plt.ylim(0, 1.1)
plt.title("5-Fold Cross Validation Accuracy (Subject Model)")
plt.ylabel("Accuracy")
plt.xlabel("Fold")
for i, v in enumerate(scores):
    plt.text(i+0.9, v+0.02, f"{v:.2f}")
plt.show()
```

## Section 7: K-Means Clustering Demonstration
Grouping teachers based on behavioral features: Punctuality, Consistency, Completeness, and Volume.


```python
# Generate synthetic teacher data
np.random.seed(42)
teachers = [f"Teacher {i}" for i in range(1, 51)]
punctuality = np.random.normal(85, 10, 50)
consistency = np.random.normal(80, 15, 50)
completeness = np.random.normal(90, 5, 50)
volume = np.random.normal(12, 4, 50)

# Cap values at 0-100
punctuality = np.clip(punctuality, 0, 100)
consistency = np.clip(consistency, 0, 100)
completeness = np.clip(completeness, 0, 100)

data_clusters = pd.DataFrame({
    'Punctuality': punctuality,
    'Consistency': consistency,
    'Completeness': completeness,
    'Volume': volume
})

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
data_clusters['Cluster'] = kmeans.fit_predict(data_clusters)

# Renaming clusters for interpretation
cluster_names = {0: "Moderate", 1: "Highly Consistent", 2: "Needs Support"}
# (Note: Actual mapping depends on centroids, but for documentation purposes we assign labels)
centroids = kmeans.cluster_centers_
# Sort clusters by average scores to assign labels correctly
avg_scores = centroids.mean(axis=1)
ordered_indices = np.argsort(avg_scores)
mapping = {ordered_indices[0]: "Needs Support", ordered_indices[1]: "Moderate", ordered_indices[2]: "Highly Consistent"}
data_clusters['Performance Group'] = data_clusters['Cluster'].map(mapping)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Punctuality', y='Consistency', hue='Performance Group', size='Volume', data=data_clusters, palette='Set1', alpha=0.7)
plt.title("Teacher Performance Clustering (K-Means)")
plt.show()
```
